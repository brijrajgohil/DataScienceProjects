```python
# Plotting for EDA (Exploratory data analysis)
# Iris flower dataset
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
# how many data-points and features?
print(iris.shape)
```

```python
# the column names?
print(iris.columns)
```

```python
# the data points for each class?
iris["species"].value_counts()
```

```python
# 2-D Scatter plot
# it's always imp to understand what the axis represent

iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
plt.show()
```

```python
# 2-D scatter plot with color coding
# it's important to understand which color represents which class
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", height=6).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
plt.show()

# It can be observed that the blue points (setosa) are quite 
# differentiable
# Where as the green and orange points (virginica, versicolor)
# are hard to differentiate based on just sepal_width and sepal_length
```

```python
# Pair-plot
# pairwise scatter plots are known as pair plots
# cannot be used in higher dimensions
sns.pairplot(iris, hue="species", height=3)
plt.show()
# diagonals are pdfs
```

```python
# Histogram, PDF, CDF
# Probability density function (PDF)
# Cumulative density function (CDF)

# 1-D Scatter plot for just one feature
# 1-D scatter plot of petal_length
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa["petal_length"]), "o")
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica["petal_length"]), "o")
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor["petal_length"]), "o")
plt.show()
# very challenging to make the sense of points
# due to overlapping
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "petal_length").add_legend()
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "petal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "sepal_length").add_legend()
plt.show()
```

```python
sns.FacetGrid(iris, hue="species", height=5).map(sns.distplot, "sepal_width").add_legend()
plt.show()
```

```python
# Plot the CDF of petal_length
counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins=20, density=True)
pdf = counts/(sum(counts))
print(pdf)
plt.plot(bin_edges[1:], pdf)

plt.show()
```

```python
counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute cdf
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

```python
# Plots of CDF of petal_length for all of the types of flowes (classes)
counts, bin_edges = np.histogram(iris_setosa["petal_length"], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute cdf
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(iris_virginica["petal_length"], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute cdf
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(iris_versicolor["petal_length"], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

# compute cdf
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

```python
# Mean, Variance and Std-dev
print("Means: ")
print(np.mean(iris_setosa['petal_length']))
# Mean with an outlier
print(np.mean(np.append(iris_setosa["petal_length"], 50)))
print(np.mean(iris_virginica['petal_length']))
print(np.mean(iris_versicolor['petal_length']))

print("\nStd-dev")
print(np.std(iris_setosa["petal_length"]))
print(np.std(iris_virginica["petal_length"]))
print(np.std(iris_versicolor["petal_length"]))
```

```python
# Medians, Quantiles, Percentiles, IQR
print("\nMedians:")
print(np.median(iris_setosa["petal_length"]))
#Median with an outlier
print(np.median(np.append(iris_setosa["petal_length"], 50)))
print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))

print("\nQuantiles: ")
print(np.percentile(iris_setosa["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_virginica["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(iris_versicolor["petal_length"], np.arange(0, 100, 25)))

print("\n90th Percentiles: ")
print(np.percentile(iris_setosa["petal_length"], 90))
print(np.percentile(iris_virginica["petal_length"], 90))
print(np.percentile(iris_versicolor["petal_length"], 90))

from statsmodels import robust
print("\nMedian Absolute Deviation")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
```

```python
# Box-plot and whiskers
sns.boxplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
# Violin plots
sns.violinplot(x="species", y="petal_length", data=iris, height=8)
plt.show()
```

```python
# 2D density plot, contour plots
sns.jointplot(x="petal_length", y="petal_width", data=iris_setosa, kind="kde")
plt.show()
```

EDA on habermans dataset<br>
Haberman dataset contains cases from a study at the Uni of Chicago's Billings hospital on the survival of patients who had undergone surgery for breast cancer.

```python
# read the haberman data
haberman = pd.read_csv('haberman.csv')
haberman.head()
```

age : is the age of the person during the operation <br>
year: patient's year of operation <br>
nodes: number of positive axillary nodes detected <br>
status : survival status 1 (survived >= 5 years) 0 (died within 5 years)

```python
# how many datapoints and features?
print(haberman.shape)
```

```python
# list column names
print(haberman.columns)
```

```python
# is the data balanced/imbalanced?
haberman["status"].value_counts()
```

It can be observed that there are 306 data points and 4 features,
and the data is imbalanced.

```python
# 2-D scatter plot betwen age and nodes
haberman.plot(kind="scatter", x="age", y="nodes")
plt.show()
```

```python
# 2-D scatter plot with color-coding for each patient based on survival
sns.set_style("whitegrid")
sns.FacetGrid(haberman, hue="status", height=4).map(plt.scatter, "age", "nodes").add_legend()
plt.show()
```

**Observation(s)**
1. It can be observed that the data is imbalanced, plus a lot of the nodes lie between 0-10.
2. Its hard to distingush berween patients.
