```python
# Binary search
import math
def binarySearch(arr, l, r, x):
    # Check base case
    if r >= l:
        mid = l + math.floor((r - l) / 2)
        
        # mid elem == x
        if arr[mid] == x:
            return mid
        
        # mid elem > x
        elif arr[mid] > x:
            return binarySearch(arr, l, mid - 1, x)
        
        else:
            return binarySearch(arr, mid + 1, r, x)
    else:
        return -1
    
l.sort()
arr = l
q = 31
binarySearch(arr, 0, len(arr) - 1, q)
```

```python
# Find elements common in two lists
l1 = list(range(100))
random.shuffle(l1)

l2 = list(range(50))
random.shuffle(l2)

cnt = 0
for i in l1:
    for j in l2:
        if i == j:
            print(i)
            cnt += 1
print("Number of common elements: ", cnt)
```

```python
# Find elements common in two lists
l1 = list(range(100))
random.shuffle(l1)

l2 = list(range(50))
random.shuffle(l2)

smallList = {}
for ele in l2:
    smallList[ele] = 1
    
cnt = 0
for i in l1:
    if smallList.get(i) != None:
        print(i)
        cnt += 1
print("Number of common elements: ", cnt)
```
