```python
# QQ Plot
import numpy as np
import pylab
import scipy.stats as stats

# N(0, 1)
std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)

# 0 to 100th percentile of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
# generate 100 samples from N(20,5)
measurements = np.random.normal(loc = 20, scale=5, size=50000)
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# generate 100 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=10000)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# load IRIS data (150 points)
from sklearn import datasets
iris = datasets.load_iris()
d = iris.data
d.shape
print(d[0])
```

```python
# sample 30 points randomly
n = 150
m = 30
p = m/n
print(p)
sampled_data = []

for i in range(0,n):
    a = random.random()
    if a <= p:
        sampled_data.append(d[i, :])
        
print(sampled_data)
```

```python
# Confidence interval
import numpy as np
import pandas as pd
from sklearn.utils import resample
import matplotlib.pyplot as plt
%matplotlib inline

# load the data
x = np.array([180,162,158,172,168,150,171,183,165,176])

# configure bootstrap
n_iterations = 1000
n_size = int(len(x))

# run bootstrap
medians = list()
for i in range(n_iterations):
    # prepare train and test sets
    s = resample(x, n_samples=n_size)
    m = np.median(s)
    medians.append(m)
    
# plot scores
plt.hist(medians)
plt.show()

# confidence intervals
alpha = 0.95
p = ((1.0 - alpha) / 2.0) * 100
lower = np.percentile(medians, p)

p = (alpha + ((1.0 - alpha) / 2.0)) * 100
upper = np.percentile(medians, p)
print("%.1f confidence interval %.1f and %.1f" % (alpha*100, lower, upper))
```

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

# generate a gaussian r.v X
x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, "norm")
```

```python
# Y ~ Cont Uni Dist(0,1)
y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, "norm")
```
